```python
from __future__ import division

import matplotlib
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import numpy as np
import random

%matplotlib inline

class Adaline:
    def __init__(self, number_of_inputs, learning_rate):        
        self.number_of_inputs = number_of_inputs
        self.learning_rate = learning_rate        
        self._randomly_initialize_weigths()

        Adaline.ONE = np.array([[1]])
        
    def _randomly_initialize_weigths(self):
        self.weights = np.reshape(np.random.rand(self.number_of_inputs+1), (self.number_of_inputs+1,1))
            
    def activate(self, input_signal):
        return self._activate(np.concatenate((Adaline.ONE, input_signal), axis=1))
        
    def _activate(self, input_signal):
        return np.dot(input_signal, self.weights)

    def train(self, training_set, number_of_epochs):
        
        self._randomly_initialize_weigths()
        
        number_of_patterns = training_set.shape[0]
        
        self._randomly_initialize_weigths()
        for i in xrange(number_of_epochs):             
            for j in np.arange(number_of_patterns):                
                input_signal = np.concatenate((Adaline.ONE, training_set[j,:-1,None].T), axis=1)
                target = training_set[j,-1]
                
                output_signal = self._activate(input_signal)
                self.weights += self.learning_rate*(target-output_signal)*input_signal.T
```

```python
# Creating the training data
np.random.seed(15)
number_training_data = 400

negative_indices = np.arange(200);
positive_indices = np.arange(200,400);

covariance = 0.09
training_set = np.random.multivariate_normal([-1,-1],covariance*np.eye(2),number_training_data)
training_set[100:200,] += 2.5
training_set[200:300,0] += 2.5
training_set[300:400,1] += 2.5

training_target = np.ones((number_training_data))
training_target[negative_indices] = -1
training_set = np.column_stack((training_set, training_target))
```

```python
adaline = Adaline(2,0.15)
print "Number of inputs: " + str(adaline.number_of_inputs)
print "Learning rate: " + str(adaline.learning_rate)

number_of_epochs = 10000
adaline.train(training_set, number_of_epochs)
```
